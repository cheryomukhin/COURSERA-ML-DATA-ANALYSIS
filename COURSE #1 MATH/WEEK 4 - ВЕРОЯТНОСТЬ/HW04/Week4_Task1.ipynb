```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
```

# Рассмотрим распределение Эрланга
Оно является частным случаем гамма-распределения (https://ru.wikipedia.org/wiki/Гамма-распределение#Моделирование_гамма-величин) при целочисленном первом параметре k.
Формулы среднего и дисперсии:
$EX = k\theta;$
$DX = k\theta^2$

```python
dim = 1000 #размерность выборки
#параметры распределения
par1 = 4. #параметр k распределения Эрланга / гамма-распределения
par2 = 1.5 #параметр 1/lambda распределения Эрланга (или theta для общего случая - Гамма распределения)
erlang_rv = sts.erlang(a=par1, loc=0., scale=par2)
sample = erlang_rv.rvs(dim) #сгенерировали выборку

#для построения далее графика функции плотности сгенерируем её значения на отрезке от 0 до 30
x = np.linspace(0,30,150) 
pdf = erlang_rv.pdf(x) 

#Вычислим среднее и дисперсию взятого распределения
M = par1 * par2 #среднее
V = par1* (par2 ** 2) #дисперсия
print M, V
```

```python
plt.hist(sample, bins = 20, normed=True) #рисуем гистограмму
plt.plot(x, pdf, alpha=0.5) #рисуем функцию плотности
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

### Рассмотрим выборки размерностью 5

```python
sample_5 = np.array([]) #зададим массив, в который далее запишем средние значения выборки размерности n
n = 5
# в цикле формируем выборки из распределения Эрланга размерности 5, считаем среднее значение каждой такой выборки
for i in range(1000):
    sample_i = erlang_rv.rvs(n)
    smean_i = sum(sample_i)/n
    sample_5 = np.append(sample_5, smean_i) #получили массив с выборочными средними

V5 = V/n #Дисперсия нормального распределения, которое приближает полученную выборку средних
sigma5 = math.sqrt(V5) #отклонение нормального распределения
norm_n5 = sts.norm(M, sigma5) #задали нормальное распределение с значениями среднего (расичтано выше) и отклонения
x = np.linspace(0,14,150)
pdf = norm_n5.pdf(x) #функция плотности нормального распределения
```

```python
plt.hist(sample_5, bins = 20, normed=True) #гистограмма выборочных средних
plt.plot(x, pdf, alpha=0.5) #функция плотности нормального распределения, которое приближает выборочные средние
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```
